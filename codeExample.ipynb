```python
def card_conv(x,r):
    d = ''
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'

    while x > 0:
        d += dchar[x%r]
        x //= r

    return d[::-1]


print(card_conv(6,2))
```

```python


def binarySearch(a: Sequence, key: Any) -> int:
    pl = 0
    pr = len(a) - 1

    while True:
        pc = (pl + pr)//2
        if(a[pc] == key): 
            return pc
        if(a[pc] < key):
            pl = pc + 1
        else:
            pr = pc -1
        if pl > pr:
            break
    return -1

list = [1,2,3,5,7,8,16,18,44,87,324]
idx = binarySearch(list,16)

print("찾는값의 인덱스 : {}, 값은{}".format(idx, list[idx]))
```

## 큐

큐에 데이터를 추가하는 작업을 `인큐`, 데이터를 꺼내는 작업을 `디큐`라고 한다. 또 데이터를 꺼내는 쪽을 `프런트`, 데이터를 넣는 쪽을 `리어`라고 한다.

### 링 버퍼로 큐 구현하기 (디큐할 때 배열 안의 원소를 옮기지 않는 큐)

```python
from typing import Any

class FixedQueue:

    class Empty(Exception):
        pass
    class Full(Exception):
        pass

    def __init__(self, capacity: int) -> None:
        self.no = 0
        self.front = 0
        self.rear = 0
        self.capacity = capacity
        self.que = [None]*capacity

    def __len__(self) -> int:
        return self.no
    
    def is_empty(self) -> bool:
        return self.no <= 0
    
    def is_full(self) -> bool:
        return self.no >= self.capacity
    
    def enque(self, x: Any) -> None:
        if self.is_full(x):
            raise FixedQueue.Full
        self.que[self.rear] = x
        self.no += 1
        if self.rear == self.capacity
        self.rear = 0

    def deque(self) -> Any:
        if self.is_empty():
            raise FixedQueue.Empty
        x = self.que[self.front]
        self.front += 1
        self.no -= 1
        if self.front == self.capacity:
            self.front = 0
        return x
    
    def peek(self) -> Any: # 맨 앞 데이터, 즉 다임 디큐에서 꺼낼 데이터를 들여다 본다.

        if self.is_empty():
            raise FixedQueue.Empty
        return self.que[self.front]
    
    def find(self, value: Any) -> Any:
        for i in range(self.no):
            idx = (i + self.front) % self.capacity
            if self.que[idx] == value:
                return idx
        return -1
    
    def count(self, value: Any) -> bool:
        count = 0
        for i in range(self.no):
            idx = (i + self.front) % self.capacity
            if self.que[idx] == value:
                count += 1

        return count
    
    def __contains__(self, value: Any) -> bool:
        return self.count(value)
    
    def clear(self) -> None:
        self.no = self.front = self.rear = 0

    def dump(self) -> None:
        if self.is_empty():
            print("큐가 비었습니다.")
        else:
            for i in range(self.no):
                print(self.que[(i + self.front) % self.capacity], end='')
            print()
```

```python
from typing import MutableSequence

def bubble_sort(a: MutableSequence) -> None:
    for i in range(len(a)-1,-1,-1):
        for j in range(i):
            if a[j] > a[j+1]:
                a[j+1], a[j] = a[j], a[j+1]


list = [3,56,123,3,55,33,4,12,56,44,427,45,32,8,76,10]


print(list)
bubble_sort(list)
print(list)
```

### 선택 정렬 (selction sort)

```python
def selection_sort(a: MutableSequence) -> None:
    n = len(a)

    for i in range(n-1):
        min_idx = i  # 정렬할 부분에서 가장 작은 원소의 인덱스
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]



list = [3,56,123,3,55,33,4,12,56,44,427,45,32,8,76,10]


print(list)
selection_sort(list)
print(list)
```

```python
def insertion_sort(a):

    for i in range(1,len(a)):
        tmp = a[i]
        for j in range(i-1,-1,-1):
            if a[j] > tmp:
                a[j+1] = a[j]
            elif a[j] <= tmp:
                a[j+1] = tmp
                break

list = [3,56,123,3,55,33,4,12,56,44,427,45,32,8,76,10]
print(len(list))
print(list[15])
print(list)
insertion_sort(list)
print(list)
```

### 빠른 정렬 (quick sort)

```python
def partition(arr, p,r):
    i = p-1
    j = p
    x = arr[r]

    while j <= r-1
        if arr[j] >= x:
            j += 1
        else:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            j += 1
    arr[i+1], arr[r] = arr[r], arr[i + 1]

    return i+1

def quick_sort(arr, p, r):
    while p <= r:
        q = partition(arr, p, r)
        quick_sort(arr, p, q-1)
        quick_sort(arr, q+1. r)
```

```python
from typing import MutableSequence

def qsort(a: MutableSequence, left: int, right: int) -> None:

    pl = left # 왼쪽 커서
    pr = right # 오른쪽 커서
    x = a[(left + right)//2] #피벗(가운데 원소)

    while pl <= pr:
        while a[pl] < x: pl += 1
        while a[pr] > x: pr -= 1
        if pl <= pr:
            a[pl], a[pr] = a[pr], a[pl]
            pl += 1
            pr -= 1

    if left < pr: qsort(a, left, pr)
    if pl < right: qsort(a, pl, right)


def quick_sort(a: MutableSequence):
    qsort(a, 0, len(a)-1)

```

### 힙 정렬(heap sort)


`Max-Heapify`

```python
def max_heapify(arr, i, heap_size):
    left_idx = 2*i + 1
    right_idx = 2*i + 2
    largest_idx = i 

    if left_idx < heap_size and arr[left_idx] > arr[largest_idx]:
        largest_idx = left_idx
    if right_idx < heap_size and arr[right_idx]> arr[largest_idx]:
        largest_idx = right_idx

    if largest_idx != i:
        arr[i], arr[largest_idx] = arr[largest_idx], arr[i]
        max_heapify(arr, largest_idx, heap_size)
```

```python
def build_max_heap(arr):
    n = len(arr)
    
    for i in range(n//2 - 1, -1, -1):
        max_heapify(arr, i, n)
```

```python
def heap_sort(arr):
    build_max_heap(arr)

    for i in range(len(arr)-1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        max_heapify(arr, 0, i)
```

### 힙의 응용 : 최대 우선순위 큐

```python
class MaxPriorityQueue:
    def __init__(self):
        self.heap = []

    def parent(self, i):
        return (i - 1) // 2

    def left_child(self, i):
        return 2 * i + 1

    def right_child(self, i):
        return 2 * i + 2

    def max_heapify(self, i, heap_size):
        left = self.left_child(i)
        right = self.right_child(i)
        largest = i

        if left < heap_size and self.heap[left] > self.heap[largest]:
            largest = left

        if right < heap_size and self.heap[right] > self.heap[largest]:
            largest = right

        if largest != i:
            self.heap[i], self.heap[largest] = self.heap[largest], self.heap[i]
            self.max_heapify(largest, heap_size)

    def build_max_heap(self):
        n = len(self.heap)
        for i in range(n // 2 - 1, -1, -1):
            self.max_heapify(i, n)

    def insert(self, key):
        self.heap.append(key)
        i = len(self.heap) - 1
        while i > 0 and self.heap[self.parent(i)] < self.heap[i]:
            self.heap[i], self.heap[self.parent(i)] = self.heap[self.parent(i)], self.heap[i]
            i = self.parent(i)

    def maximum(self):
        if len(self.heap) == 0:
            raise Exception("Queue is empty")
        return self.heap[0]

    def extract_max(self):
        if len(self.heap) == 0:
            raise Exception("Queue is empty")
        max_val = self.heap[0]
        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self.max_heapify(0, len(self.heap))
        return max_val

    def is_empty(self):
        return len(self.heap) == 0
```

```python
def counting_sort(arr):
    # Find the maximum element in the array
    max_elem = max(arr)

    # Create a counting array of size max_elem+1 and initialize all elements to 0
    count_arr = [0] * (max_elem+1)

    # Count the frequency of each element in the array
    for elem in arr:
        count_arr[elem] += 1

    # Calculate the cumulative sum of the count array
    for i in range(1, len(count_arr)):
        count_arr[i] += count_arr[i-1]

    # Create a sorted array of the same length as the input array
    sorted_arr = [0] * len(arr)

    # Place each element from the input array into its correct position in the sorted array
    for elem in reversed(arr):
        sorted_arr[count_arr[elem]-1] = elem
        count_arr[elem] -= 1

    return sorted_arr
```

- 시간복잡도 : O(n+k)
- k가 클 경우 비실용적
- stable 정렬 알고리즘 : 입력에 동일한 값이 있을때 입력에 먼저 나오는 값이 출력에서도 먼저 나온다



### Radix Sort
- n 개의 d자리 정수들
- 가장 낮은 자리수부터 정렬

```python
def radix_sort(arr):
    # Find the maximum number to know the number of digits
    max_num = max(arr)
    
    # Do counting sort for every digit
    exp = 1
    while max_num // exp > 0:
        counting_sort(arr, exp)
        exp *= 10
        
def counting_sort(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10
    
    # Count occurrences of each digit in arr
    for i in range(n):
        digit = (arr[i] // exp) % 10
        count[digit] += 1
    
    # Calculate cumulative count for each digit
    for i in range(1, 10):
        count[i] += count[i - 1]
    
    # Build the output array
    i = n - 1
    while i >= 0:
        digit = (arr[i] // exp) % 10
        output[count[digit] - 1] = arr[i]
        count[digit] -= 1
        i -= 1
    
    # Copy the output array to the input array
    for i in range(n):
        arr[i] = output[i]
```

## 검색트리 (Search tree)
- Dynamic set을 배열이나 연결리스트로 구현하여 search, insert, delete 연산을 했을 때 어떠한 경우라도 하나는 O(n)의 시간복잡도를 가짐
- Dynamic set을 트리의 형태로 구현 (search, insert, delete 연산을 효율적으로 하기 위함) O(h) (h:트리의 높이)
- 일반적으로 SEARCH, INSERT, DELETE 연산이 트리의 높이에 비례하는 시간복잡도를 가짐
- 이진검색트리, 레드-블랙 트리, B-트리

- 노드의 개수가 n개면 트리는 항상 n-1개의 링크를 가진다.
- 트리에서 루트에서 어떤 노드로 가는 경로는 유일하다. 또한 임의의 두 노드간의 경로도 유일하다.


### 트리의 검색
어느 시점에 노드를 방문하는지에 따라 결정됨
- 전위 순회(preorder) : 노드 방문 -> 왼쪽 자식 -> 오른쪽 자식
- 중위 순위(inorder) : 왼쪽 자식 -> 노드 방문 -> 오른쪽 자식
- 후위 순위(postorder) : 왼쪽 자식 -> 오른쪽 자식 -> 노드 방문 


### 이진 트리 (binary tree)

- 이진 트리에서 각 노드는 최대 2개의 자식을 가진다.
- 각각의 자식 노드는 자신이 부모의 왼쪽 자식인지 오른쪽 자신인지가 지정된다.

### 이진 검색 트리
- 왼쪽 서브트리의 노드의 키값은 자신의 노드 키값보다 작거나 같다.
- 오른쪽 서브트리의 노드의 키값은 자신의 노드 키값보다 크거나 같다.

#### 최솟값 / 최댓값
- 최소값은 항상 가장 왼쪽 노드에 존재한다.
- 최대값은 항상 가장 오른쪽 노드에 존재한다.

#### Successor
- 노드 x의 successor란 key[x]보다 크면서 가장 작은 키를 가진 노드
- 모든 키들이 서로 다르다고 가정
  
- 3가지 경우
    - 노드 x의 오른쪽 부트리가 존재할 경우, 오른쪽 부트리의 최소값.
    - 오른쪽 부트리가 없는 경우, 어떤 노드 y의 왼쪽 부트리의 최대값이 x가 되는 그런 노드 y가 x의 successor
        - 부모를 따라 루트까지 올라가면서 처음으로 누군가의 왼쪽 자식이 되는 노드
    - 그런 노드 y가 존재하지 않을 경우 successor가 존재하지 않음 (즉, x가 최대값)

#### Predecessor
- key[x]보다 작으면서 가장 큰 키를 가진 노드

#### SEARCH

1. 루트에 주목한다. 여기서 주목하는 노드를 p라고 한다.
2. p가 None이면 검색을 실패하고 종료한다.
3. 검색하는 key와 주목 노드 p의 키를 비교한다.
    - key = p : 검색을 성공하고 종료한다.
    - key < p : 주목 노드를 왼쪽 자식 노드로 옮긴다.
    - key > p : 주목 노드를 오른쪽 자식 노드로 옮긴다.
4. 2번 과정으로 되돌아 간다.

#### INSERT

1. 루트에 주목한다. 여기서 주목하는 노드를 node라고 한다.
2. 삽입하는 key와 주목 노드 node의 키를 비교한다.
    - key = node인 경우 : 삽입을 실패하고 종료한다.
    - key < node인 경우:
        - 왼쪽 자식 노드가 없으면, 그 자리에 노드를 삽입하고 종료한다.
        - 왼쪽 자식 노드가 있으면, 주목 노드를 왼쪽 자식 노드로 옮긴다.
    - key > node:
        - 오른쪽 자식 노드가 없으면, 그 자리에 노드를 삽입하고 종료한다.
        - 오른쪽 자식 노드가 있으면, 주목 노드를 오른쪽 자식 노드로 옮긴다.
3. 2번 과정으로 되돌아간다.

#### DELETE
- `case 1: 자식노드가 없는 경우 -> 그냥 삭제`
    1. 삭제할 노드가 부모 노드의 왼쪽 자식이면, 부모의 왼쪽 포인터를 None으로 한다.
    2. 삭제할 노드가 부모 노드의 오른쪽 자식이면, 부모의 오른쪽 포인터를 None으로 한다.
- `case 2: 자식노드가 하나인 경우 -> 자신의 자식노드를 원래 자신의 위치로`
    1. 삭제할 노드가 부모 노드의 왼쪽 자식인 경우: 부모의 왼쪽 포인터가 삭제할 노드의 자식을 가리키도록 업데이트
    2. 삭제할 노드가 부모 노드이ㅡ 오른쪽 자식의 경우: 부모의 오른쪽 포인터가 삭제할 노드의 자식을 가리키도록 업데이트
- `case 3: 자식노드가 2개인 경우 -> 삭제하려는 노드의 successor, successor 노드는 자식의 개수가 최대 1개(case 1, case2 에 해당)`
    1. 삭제할 노드의 왼쪽 서브트리에서 키값이 가장 큰 노드를 검색하나.
    2. 검색한 노드를 삭제 위치로 옮긴다. 즉, 검색한 노드의 데이터를 삭제할 노드 위치에 복사한다.
    3. 옮긴 노드를 삭제한다. 이때 자식 노드의 개수에 따라 다음을 수행한다.
        - 옮기 노드에 자식이 없다 : case 1
        - 옮긴 노드에 자식이 1개만 있으면 : case 2


#### dump()
모든 노드를 오름차순으로 출력하는 함수, 스캔은 중위 순회의 깊이 우선 검색으로 한다.

```python
from __future__ import annotations
from typing import Any, Type

class Node:
    def __init__(self, key: Any, value: Any, left: Node = None, right: Node = None):
        self.key = key
        self.value = value
        self.left = left
        self.right = right

class BinarySearchTree:
    def __init__(self):
        self.root = None
    
    def search(self, key:Any)-> Any:
        p = self.root
        while True:
            if p is None:
                return None
            if key == p.key:
                return p.value
            elif key < p.key:
                p = p.left
            else:
                p = p.right

    # 노드를 삽입할 때에는 검색할 때와 마찬가지로 먼저 삽입할 위치를 찾아낸 뒤에 수행해야 한다.
    def add(self, key: Any, value: Any) -> bool:

        def add_node(node: Node, key: Any, value: Any) -> None:
            if key == node.key:
                return False
            elif key < node.key:
                if node.left is None:
                    node.left = Node(key, value, None, None)
                else:
                    add_node(node.left, key, value)
            else:
                if node.right is None:
                    node.right = Node(key, value, None, None)
                else:
                    add_node(node.right, key, value)
            return True
        
        if self.root is None:
            self.root = Node(key, value, None, None)
            return True
        else:
            return add_node(self.root, key, value)



    def delete(self, key: Any) -> bool:
        p = self.root # 스캔중인 노드
        parent = None # 스캔 중인 노드의 부모
        is_left_child = True # p는 parent의 왼쪽 자식 노드인지 확인 -> 부모 노드의 어떤 포인터에 하위 노드를 연결해야하는지 알아야 하므로 필요하다.

        while True: # 삭제할 노드를 검색한다. 검색에 성공하면 p는 삭제할 노드를 가리키고 parent는 그 노드의 부모 노드를 가리킨다.
            if p is None:
                return False
            
            if key == p.key:
                break

            else:
                parent = p
                if key < p.key:
                    is_left_child = True
                    p = p.left
                else:
                    is_left_child = False
                    p = p.right
    
        if p.left is None: # 스캔중인 노드의 왼쪽 자식이 없으면 
            if p is self.root: # 스캔한 노드가 root 노드이면 
                self.root = p.right
            elif is_left_child:
                parent.left = p.right
            else:
                parent.right = p.right # 스캔중이 노드가 오른쪽 자식이였을 경우 그대로 오른쪽 하위 노드를 연결
        
        elif p.right is None:
            if p is self.root:
                self.root = p.lef
            elif is_left_child:
                parent.left = p.left
            else:
                parent.right = p.left

        else:  #자식 노드가 2개인 노드를 삭제하는 경우
            parent = p
            left = p.left # 왼쪽 서브트리 안에서 가장 큰 노드가 들어올 변수
            is_left_child = True

            while left.right is not None: # 가장 큰 노드 left를 검색 -> 오른쪽 자식이 없을때 까지
                parent = left
                left = left.right
                is_left_child = False

            p.key = left.key
            p.value = left.value

            if is_left_child:
                parent.left = left.left
            else:
                parent.right = left.left

            return True
        
    #모든 노드를 오름차순으로 출력하는 dump()함수
    def dump(self, reverse = False )-> None:

        def print_subtree(node: Node):

            if node is not None:
                print_subtree(node.left)
                print(f'{node.key}  {node.value}')
                print_subtree(node.right)

            print_subtree(self.root)
        
        def print_subtree_rev(node: Node):

            if node is not None:
                print_subtree_rev(node.right)
                print(f'{node.key} {node.value}')
                print_subtree_rev(node.left)

        print_subtree_rev(self.root) if reverse else print_subtree(self.root)

    def min_key(self) -> Any:
        if self.root is None:
            return None
        
        p = self.root

        while p.left is not None:
            p = p.left

        return p.key
    
    def max_key(self) -> Any:
        if self.root is None:
            return None
        
        p = self.root
        while p.right is not None:
            p = p.right
        return p.key
```

```python
class TreeNode:
    def __init__(self, value: Any, left: TreeNode = None, right: TreeNode = None):
        self.value = value
        self.left = left
        self.right = right

class BinarySearchTree:
    def __init__(self):
        self.root = None
    
    def search(self, value: Any) -> Any:
        p = self.root
        while True:
            if p is None:
                return None
            if value == p.value:
                return p.value
            elif value < p.value:
                p = p.left
            else:
                p = p.right

    def add(self, value: Any) -> bool:

        def add_node(node: TreeNode, value: Any) -> None:
            if value == node.value:
                return False
            elif value < node.value:
                if node.left is None:
                    node.left = TreeNode(value, None, None)
                else:
                    add_node(node.left, value)
            else:
                if node.right is None:
                    node.right = TreeNode(value, None, None)
                else:
                    add_node(node.right, value)
            return True
        
        if self.root is None:
            self.root = TreeNode(value, None, None)
            return True
        else:
            return add_node(self.root, value)

    def delete(self, value: Any) -> bool:
        p = self.root
        parent = None
        is_left_child = True

        while True:
            if p is None:
                return False
            
            if value == p.value:
                break

            else:
                parent = p
                if value < p.value:
                    is_left_child = True
                    p = p.left
                else:
                    is_left_child = False
                    p = p.right
    
        if p.left is None:
            if p is self.root:
                self.root = p.right
            elif is_left_child:
                parent.left = p.right
            else:
                parent.right = p.right
        
        elif p.right is None:
            if p is self.root:
                self.root = p.left
            elif is_left_child:
                parent.left = p.left
            else:
                parent.right = p.left

        else:
            parent = p
            left = p.left
            is_left_child = True

            while left.right is not None:
                parent = left
                left = left.right
                is_left_child = False

            p.value = left.value

            if is_left_child:
                parent.left = left.left
            else:
                parent.right = left.left

            return True
        
    def dump(self, reverse = False ) -> None:

        def print_subtree(node: TreeNode):

            if node is not None:
                print_subtree(node.left)
                print(node.value)
                print_subtree(node.right)

            print_subtree(self.root)
        
        def print_subtree_rev(node: TreeNode):

            if node is not None:
                print_subtree_rev(node.right)
                print(node.value)
                print_subtree_rev(node.left)

        print_subtree_rev(self.root) if reverse else print_subtree(self.root)

    def min_value(self) -> Any:
        if self.root is None:
            return None
        
        p = self.root

        while p.left is not None:
            p = p.left

        return p.value
    
    def max_value(self) -> Any:
        if self.root is None:
            return None
        
        p = self.root
        while p.right is not None:
            p = p.right
        return
```

### 문제 1\) 정렬된 배열의 이진 탐색 트리 변환


```python
def insert_into_bst(arr):
    if not arr:
        return None

    mid = len(arr) // 2
    root = TreeNode(arr[mid])
    root.left = insert_into_bst(arr[:mid])
    root.right = insert_into_bst(arr[mid+1:])
    return root
```

### 문제2\) 이진 검색 트리를 순회하면 값출력하기

```python
def preorder_traversal(self, node: TreeNode = None) -> None:
    if node is None:
        node = self.root
    if node is not None:
        print(node.value)
        self.preorder_traversal(node.left)
        self.preorder_traversal(node.right)

def inorder_traversal(self, node: TreeNode = None) -> None:
    if node is None:
        node = self.root
    if node is not None:
        self.inorder_traversal(node.left)
        print(node.value)
        self.inorder_traversal(node.right)

def postorder_traversal(self, node: TreeNode = None) -> None:
    if node is None:
        node = self.root
    if node is not None:
        self.postorder_traversal(node.left)
        self.postorder_traversal(node.right)
        print(node.value)
```

### 문제3\) 깊이 구하기

```python
def depth(self, node: TreeNode = None) -> int:
        if node is None:
            node = self.root
        if node is None:
            return 0
        return 1 + max(self.depth(node.left), self.depth(node.right))
```
